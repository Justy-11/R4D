```python
import tensorflow as tf
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
import shutil

# Set the root directory
root_dir = 'D:\JATHURSH\SEMI-06\project\mrlEyes_2018_01\separated\mrlEyes_2018_01'

# Loop through the folders in the root directory
for folder in os.listdir(root_dir):
  # Check if the item is a folder
  if os.path.isdir(os.path.join(root_dir, folder)):
    # Set the directory containing the images
    current_img_dir = os.path.join(root_dir, folder)
    
    # Set the directories where the images will be moved
    open_eyes = 'D:\JATHURSH\SEMI-06\project\mrlEyes_2018_01\separated\open_eyes'
    close_eyes = 'D:\JATHURSH\SEMI-06\project\mrlEyes_2018_01\separated\close_eyes'

    # Iterate over the files in the image directory
    for filename in os.listdir(current_img_dir):
        # Check if the file is an image file
        if filename.endswith('.jpg') or filename.endswith('.png'):
            # Check the 17th character in the file name
            if filename[16] == '0':
                # Move the file to the close eyes directory
                shutil.move(os.path.join(current_img_dir, filename), close_eyes)
            else:
                # Move the file to the open eyes directory
                shutil.move(os.path.join(current_img_dir, filename), open_eyes)
```
